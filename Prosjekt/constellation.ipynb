```python
import matplotlib.pyplot as plt

# Konstellasjonspunktene
c_p = [
    0.707 + 0.707j,
    -0.707 + 0.707j,
    -0.707 - 0.707j,
    0.707 - 0.707j
]
mapping = [0,1,2,3]

circle = plt.Circle((0, 0), 1, color='k', fill=False, lw=0.5, ls='--')
# lag en firkant mellom de 4 c_p punktene
square = plt.Rectangle((-0.707, -0.707), 1.414, 1.414, color='k', fill=False, lw=0.5, ls='--')
colors = ['tab:orange', 'tab:blue', 'tab:green', 'tab:red']

# Generer konstellasjonsdiagrammet
plt.figure(figsize=(8, 8))
# Tegn piler for aksene
plt.arrow(-1.1, 0, 2.2, 0, head_width=0.075, head_length=0.075, fc='black', ec='black', zorder=2)  # Reell akse
plt.arrow(0, -1.1, 0, 2.2, head_width=0.075, head_length=0.075, fc='black', ec='black', zorder=2)  # Imaginær akse
plt.xlim(-1.1, 1.175)
plt.ylim(-1.1, 1.175)
for i, p in enumerate(c_p):
    plt.scatter(p.real, p.imag, color=colors[i], s=100,zorder=2)
    plt.text(p.real*1.1, p.imag*1.1, f'{i}', fontsize=16, ha='center', va='center', color=colors[i])

# Legg til sirkel og firkant
plt.gca().add_artist(circle)
plt.gca().add_artist(square)
# Fjern aksene
plt.gca().set_xticks([])
plt.gca().set_yticks([])
plt.gca().spines['left'].set_color('none')
plt.gca().spines['bottom'].set_color('none')
plt.gca().spines['right'].set_color('none')
plt.gca().spines['top'].set_color('none')
plt.gca().set_aspect('equal', adjustable='box')
plt.text(1.1, -0.1, "In-Phase", fontsize=16, ha='left', va='center')
plt.text(0.05, 1.15, "Quadrature", fontsize=16, ha='left', va='center')
# legg til ticks direkte på pilene, +-1 og +-j
plt.text(1.0, 0.03, r'$1$', fontsize=16, ha='left', va='center')
plt.text(-1.01, 0.03, r'$-1$', fontsize=16, ha='right', va='center')
plt.text(0.03, 1.0, r'$j$', fontsize=16, ha='left', va='bottom')
plt.text(0.03, -1.01, r'$-j$', fontsize=16, ha='left', va='top')
# Legg til ticks på pilene, sqrt(2)/2 og -sqrt(2)/2
plt.text(0.707+0.02, 0.075, r'$\frac{\sqrt{2}}{2}$', fontsize=16, ha='left', va='center')
plt.text(-0.707-0.02, 0.075, r'$-\frac{\sqrt{2}}{2}$', fontsize=16, ha='right', va='center')
plt.text(0.02, 0.707+0.02, r'$j\frac{\sqrt{2}}{2}$', fontsize=16, ha='left', va='bottom')
plt.text(0.02, -0.707-0.02, r'$-j\frac{\sqrt{2}}{2}$', fontsize=16, ha='left', va='top')

# sett bakgrunn til gjennomsiktig
plt.gcf().patch.set_alpha(0.0)
plt.gca().patch.set_alpha(0.0)
plt.savefig("qpsk_constellation.pdf", bbox_inches='tight', transparent=True)
plt.show()
```
