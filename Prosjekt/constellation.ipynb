```python
import matplotlib.pyplot as plt

# Konstellasjonspunktene
c_p = [
    0.707 + 0.707j,
    -0.707 + 0.707j,
    -0.707 - 0.707j,
    0.707 - 0.707j
]
mapping = [0,1,2,3]

circle = plt.Circle((0, 0), 1, color='k', fill=False, lw=0.5, ls='--')
colors = ['tab:orange', 'tab:blue', 'tab:green', 'tab:red']
# Generer konstellasjonsdiagrammet
plt.figure(figsize=(8, 8))
plt.title("Konstellasjonsdiagram")
plt.xlabel("Reell del")
plt.ylabel("Imaginær del")
plt.grid(True, which='both', linestyle='--', lw=0.5)
# Tegn piler for aksene
plt.arrow(-1.1, 0, 2.2, 0, head_width=0.05, head_length=0.1, fc='black', ec='black', zorder=2)  # Reell akse
plt.arrow(0, -1.1, 0, 2.2, head_width=0.05, head_length=0.1, fc='black', ec='black', zorder=2)  # Imaginær akse
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
for i, p in enumerate(c_p):
    plt.scatter(p.real, p.imag, color=colors[i], s=100,zorder=2)
    plt.text(p.real*1.1, p.imag*1.1, rf'{mapping[i]:02d}$\rightarrow${i}', fontsize=12, ha='center', va='center', color=colors[i])

# Legg til sirkel
plt.gca().add_artist(circle)
plt.show()
```
